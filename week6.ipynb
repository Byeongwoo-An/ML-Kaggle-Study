```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
X = 2 * np.random.rand(100,1)
y = 6 +4 * X+np.random.randn(100,1)
plt.scatter(X,y)
```

```python
mean_X = np.mean(X)
mean_y = np.mean(y)
```

```python
b1 = np.sum((X-mean_X)*(y-mean_y))/np.sum(np.square(X-mean_X))
b0 = mean_y - (b1*mean_X)
```

```python
b0 , b1
```

```python
b_pred = b0 + (b1*X)
```

```python
plt.scatter(X,y)
plt.plot(X,b_pred)
```

```python
#w_pred = (4.022*X) + 6.162
w_pred = (4.028*X) + 6.156
```

```python
plt.scatter(X,y)
plt.plot(X,w_pred)
```

```python
b_mse = np.sum(np.square(y-b_pred))/100
w_mse = np.sum(np.square(y-w_pred))/100
b_mse, w_mse
```
