```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data = data[['id','title','crew']]
```

```python
data.head()
```

```python
import ast
```

```python
data.crew=data.crew.fillna('[{}]')
crewList=[]
for index,row in data.crew.iteritems():
    crewStr=''
    listofDict=ast.literal_eval(row)
    for dic in listofDict:
        
        if('Directing' in dic.values()):
            crewStr=crewStr+','+dic['name'] 
    crewStr=crewStr.strip(',')
    crewList.append(crewStr)
    
tempDF=pd.DataFrame(crewList,columns=['crew'])
data.crew=tempDF['crew']
```

```python
data.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data['crew'] = data['crew'].str.split(',')
data.head()
```

```python
data2 = data.copy()
```

```python
one_hot = MultiLabelBinarizer()
df = data.join(pd.DataFrame(one_hot.fit_transform(data.pop('crew')),columns=one_hot.classes_,index=data.index))
df.head()
```

```python
df2 = df.iloc[:,5:]
df2
```

```python
ibc = cosine_similarity(df2)
ibc = pd.DataFrame(data=ibc, index=df2.index, columns=df2.index)
ibc.head()
```

```python
movie_name = input()
i = int(df[df['title']==movie_name].index.values)
rank = ibc.loc[i].rank(method='min',ascending=False)
rank_df = rank.to_frame()
rank_df.columns=['rank']
rank_df.sort_values(by=['rank'],inplace=True)
rank_df = rank_df[:5]
crew = pd.merge(rank_df, data2, left_index=True, right_index=True, how='left')
crew
```
